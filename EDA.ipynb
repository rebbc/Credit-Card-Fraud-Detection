```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

<b>EDA</b>

```python
df = pd.read_csv('creditcard_2023.csv') #Carga del dataset
pd.DataFrame(df)
```

```python
df.info() #Verifico tipos de datos y valores nulos.
```

```python
plt.figure(figsize=(6, 4))

sns.countplot(df, palette= 'mako',x='Class')
plt.title('Cantidad de transacciones segun clase', size=16)
plt.xlabel('Clase')
plt.ylabel('Total Transacciones')


plt.show()
```

En este gr√°fico de barras observamos que hay la misma cantidad de operaciones fraudulentas (1) y no fraudulentas (0). Esto nos indica que las clases no est√°n desbalanceadas, por lo tanto, no es necesario aplicar t√©cnicas para balancear el dataset. 

```python
matriz_correlacion = df.corr() #Se define la matriz de correlaci√≥n entre las variables. 
```

```python
plt.figure(figsize=(18, 12))

sns.heatmap(matriz_correlacion, annot=True, fmt= '.2f', cmap='mako')

plt.title('Mapa de calor de las correlaciones', size=16)

plt.show()
```

En una matriz de correlaci√≥n, cada celda muestra el valor de correlaci√≥n entre dos variables. Los valores van desde -1 (correlaci√≥n negativa perfecta) hasta +1 (correlaci√≥n positiva perfecta). Los valores cercanos a cero indican una correlaci√≥n muy d√©bil o inexistente.
<br>
En el gr√°fico, los tonos m√°s claros indican una correlacion positiva y los m√°s oscuros, una correlaci√≥n negativa. Los valores cercanos a cero, se ven en celeste. 
<br><br>
Esta matriz muestra varias correlaciones significativas entre las variables, lo cual es √∫til para el an√°lisis. Variables con correlaciones altas con Class pueden ser especialmente relevantes para un modelo de detecci√≥n de fraude. En el gr√°fico, V4 y Class tienen una correlaci√≥n positiva, lo cual indica que V4 podr√≠a ser un buen predictor de la clase.

```python
plt.figure(figsize=(8, 6))

sns.boxplot(df, x='Amount', palette='mako')


plt.title('Monto de las transacciones', size=16)
plt.xlabel('Monto')
plt.show()
```

En este gr√°fico, descartamos la presencia de outliers en los montos de las transacciones. 

```python
variables = df.loc[:, 'V1':'V12']

plt.figure(figsize=(15, 6))

sns.boxplot(data=variables)

plt.title('Boxplots de V1 a V12 para verificar outliers', size=16)
plt.xticks(rotation=0) 
plt.show()
```

Tomamos algunas variables para analizar si tienen valores at√≠picos. <br><br> 
En general, parece que la mayor√≠a de las variables tienen una mediana cercana a cero. Los puntos fuera de los boxplot son considerados valores at√≠picos (outliers). Estos valores se encuentran a una distancia significativa de la mediana. Algunas variables muestran una dispersi√≥n alrededor de la mediana, mientras que otras parecen estar m√°s sesgadas. Las cajas m√°s alargadas indican mayor dispersi√≥n en los datos. La presencia de tantos outliers es relevante, ya que en problemas de detecci√≥n de fraude, estos valores at√≠picos pueden representar casos de fraude.<br><br>
Estos valores at√≠picos se conservar√°n porque pueden ser indicativos de diferentes patrones de comportamiento, y podr√≠an ser √∫tiles al construir un modelo de predicci√≥n.
