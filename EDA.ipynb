```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

<b>EDA</b>

```python
df = pd.read_csv('creditcard_2023.csv') #Carga del dataset
pd.DataFrame(df)
```

```python
df.info() #Verifico tipos de datos y valores nulos.
```

```python
plt.figure(figsize=(6, 4))

sns.countplot(df, palette= 'mako',x='Class')
plt.title('Cantidad de transacciones segun clase', size=16)
plt.xlabel('Clase')
plt.ylabel('Total Transacciones')


plt.show()
```

En este gráfico de barras observamos que hay la misma cantidad de operaciones fraudulentas (1) y no fraudulentas (0). Esto nos indica que las clases no están desbalanceadas, por lo tanto, no es necesario aplicar técnicas para balancear el dataset. 

```python
matriz_correlacion = df.corr() #Se define la matriz de correlación entre las variables. 
```

```python
plt.figure(figsize=(18, 12))

sns.heatmap(matriz_correlacion, annot=True, fmt= '.2f', cmap='mako')

plt.title('Mapa de calor de las correlaciones', size=16)

plt.show()
```

En una matriz de correlación, cada celda muestra el valor de correlación entre dos variables. Los valores van desde -1 (correlación negativa perfecta) hasta +1 (correlación positiva perfecta). Los valores cercanos a cero indican una correlación muy débil o inexistente.
<br>
En el gráfico, los tonos más claros indican una correlacion positiva y los más oscuros, una correlación negativa. Los valores cercanos a cero, se ven en celeste. 
<br><br>
Esta matriz muestra varias correlaciones significativas entre las variables, lo cual es útil para el análisis. Variables con correlaciones altas con Class pueden ser especialmente relevantes para un modelo de detección de fraude. En el gráfico, V4 y Class tienen una correlación positiva, lo cual indica que V4 podría ser un buen predictor de la clase.

```python
plt.figure(figsize=(8, 6))

sns.boxplot(df, x='Amount', palette='mako')


plt.title('Monto de las transacciones', size=16)
plt.xlabel('Monto')
plt.show()
```

En este gráfico, descartamos la presencia de outliers en los montos de las transacciones. 

```python
variables = df.loc[:, 'V1':'V12']

plt.figure(figsize=(15, 6))

sns.boxplot(data=variables)

plt.title('Boxplots de V1 a V12 para verificar outliers', size=16)
plt.xticks(rotation=0) 
plt.show()
```

Tomamos algunas variables para analizar si tienen valores atípicos. <br><br> 
En general, parece que la mayoría de las variables tienen una mediana cercana a cero. Los puntos fuera de los boxplot son considerados valores atípicos (outliers). Estos valores se encuentran a una distancia significativa de la mediana. Algunas variables muestran una dispersión alrededor de la mediana, mientras que otras parecen estar más sesgadas. Las cajas más alargadas indican mayor dispersión en los datos. La presencia de tantos outliers es relevante, ya que en problemas de detección de fraude, estos valores atípicos pueden representar casos de fraude.<br><br>
Estos valores atípicos se conservarán porque pueden ser indicativos de diferentes patrones de comportamiento, y podrían ser útiles al construir un modelo de predicción.
